```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as web
plt.style.use("seaborn")
```

# Very simple trading technique

```python
#Loading the stock price information for Apple
data = web.DataReader("GBPUSD=X","yahoo")
```

```python
data.head(1)
```

```python
#Simple Price Action
plt.figure(figsize=(12,8))
data["Adj Close"].plot()
```

```python
#Creating moving averages
data["14 WMA"] = data["Adj Close"].ewm(span=14).mean()
data["30 WMA"] = data["Adj Close"].ewm(span=30).mean()
```

```python
#Plotting them out
plt.figure(figsize=(12,8))
data["2020"]["Adj Close"].plot(label="Price Action",alpha=0.5)
data["2020"]["14 WMA"].plot(label="14 WMA",style="--")
data["2020"]["30 WMA"].plot(label="30 WMA")
plt.tight_layout()
plt.legend()
```

```python
#This strategy is based on Weighted Moving Averages - Their crossing indicates opportunity for buy or for sell
```
